#PCA for Free chart class logs at test


```python
# Free chart class logs at test metrics list
sheet_4.columns
```

```python
features= ['HT Effort', 'TNofNewData', 'TNofAssert', 'TLOC',
       'TINVOK', 'LCOM', 'LCOM*', 'CBO', 'FO', 'DIT', 'NOC', 'LOC', 'NOA',
       'NOO', 'RFC', 'WMPC']
```

```python
X = sheet_4.loc[:, features].dropna()
%matplotlib inline
```

```python
# I used PCA to predict the 'HT Effort' 
target='HT Effort'
y = X.loc[:,[target]].values
X = StandardScaler().fit_transform(X)
pd.DataFrame(data = X, columns = features).head()
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
principalDf.head(5)
```

```python
sheet=sheet_4.dropna()
finalDf = pd.concat([principalDf.reset_index(), sheet[[target]].reset_index()], axis = 1)
finalDf
```

```python
# preparing samples from the whole 'HT Effort' value's range
Low = sheet['HT Effort'].min()
Mean = sheet['HT Effort'].mean()
High = sheet['HT Effort'].max()
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = [Low, Mean, High]
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['HT Effort'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

#Logistic regression

```python
# Regression for appache test
X = sheet_2['NOO'].values.reshape(-1, 1)
Y= sheet_2['HT Effort'].values.reshape(-1, 1)
X=X.astype(int)
Y=Y.astype(int)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
# Regression for Free chart test
X = sheet_4['FO'].values.reshape(-1, 1)
Y= sheet_4['HT Effort'].values.reshape(-1, 1)
X=X.astype(int)
Y=Y.astype(int)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
cm = metrics.confusion_matrix(y_test, predictions)
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);
plt.savefig('toy_Digits_ConfusionSeabornCodementor.png')
plt.show()
```
